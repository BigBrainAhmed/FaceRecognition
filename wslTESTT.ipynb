```python
import os
import cv2

# Specify the path to your folder
folder_path = "C:\\Users\\Ahmed\\Documents\\GitHub\\FaceRecognition\\samplePhotos"

# Get a list of all files in the folder
file_names = os.listdir(folder_path)

# Loop through all files
for file_name in file_names:
    # Construct the full path to the file
    file_path = os.path.join(folder_path, file_name)
    
    # Check if the path is a file (and not a directory)
    if os.path.isfile(file_path):
        # Read the image
        img = cv2.imread(file_path)
        
        # Check if image is read properly
        if img is None:
            print(f"Could not read image: {file_name}")
            continue
        
        # Resize the image
        new_img = cv2.resize(img, (420, 270))
        
        # Crop the image
        img = new_img[20:20+250, 75:75+250, :]
        
        # Save the processed image to a new file (or overwrite the existing file)
        output_path = os.path.join("C:\\Users\\Ahmed\\Documents\\GitHub\\FaceRecognition\\newPhotos", f"processed_{file_name}")
        cv2.imwrite(output_path, img)
```
